```python
import pickle

filename_cluster = 'clusters_8.pkl'

with open(filename_cluster, 'rb') as handle:
    cluster_dict = pickle.load(handle)

print(cluster_dict)
```

```python
import pandas as pd
l = [[] for i in range(20500)]
cluster_df = pd.DataFrame(l)
cluster_df['value'] = ''
cluster_df['cluster'] = ''
```

```python
for c, p in cluster_dict.items():
  for elem in p:
    i = elem[0]
    v = elem[1]
    cluster_df['value'][i] = v
    cluster_df['cluster'][i] = c
```

```python
c_dict = {i:[] for i in range(8)}
c_dict
```

```python
for i in minimal_df.index:
    c = minimal_df['cluster'][i]
    v = minimal_df['value'][i]
    c_dict[c].append(v)
```

```python
c_dict
```

```python
import numpy as np 
th_list = []
for k, v in c_dict.items():
    print(k)
    th = np.percentile(np.array(v), 25)
    th_list.append(th)     
```

```python
th_list
```

```python
sorted(range(5), key=int, reverse=True)
```

```python
minimal_df['class'] = ''
for i in range(20500):
    v = minimal_df['value'][i]
    c = minimal_df['cluster'][i]
    th = th_list[c]
    if v > th:
        minimal_df['class'][i] = c
    else:
        minimal_df['class'][i] = 8
```

```python
minimal_df.head()
```

```python
import os
import numpy as np
from PIL import Image

df = minimal_df

# Create directories for each unique class if they don't exist
for class_name in df['class'].unique():
    os.makedirs(str(class_name), exist_ok=True)

# Convert numpy arrays to grayscale PNG images and save them
for idx, row in df.iterrows():
    # Convert numpy array to grayscale image
    img_data = (row['img'] * 255).astype(np.uint8)
    img = Image.fromarray(img_data, 'L')
    
    # Save the image to the corresponding folder
    img_path = os.path.join(str(row['class']), f"{idx}.png")
    img.save(img_path)
    
print("Done")
```
